## Creating an API for Stable Diffusion

```python
from fastapi.responses import JSONResponse, HTMLResponse, FileResponse
from fastapi import FastAPI, HTTPException, Request
from pydantic import BaseModel
import uvicorn
import asyncio
import nest_asyncio
from fastapi.testclient import TestClient
import os
from datetime import datetime
import json
import threading
from pyngrok import ngrok
```

```python
app = FastAPI()

class StableDiffusionInput(BaseModel):
    prompt: str

@app.post("/generate")
async def generate_image(input: StableDiffusionInput = 'a photo of a labrador astronaut sitting with a horse on mars'):
    try:
        image = pipe(input.prompt).images[0]

        #save image on server
        os.makedirs('responses_server', exist_ok=True)

        # Generate a unique filename based on the current timestamp
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        base_filename = f"response_{timestamp}"
        image_filename = f"responses_server/{base_filename}.png"
        image.save(image_filename)

        # Convert PIL Image to base64 string for transfer
        buffered = io.BytesIO()
        image.save(buffered, format="PNG")
        img_str = base64.b64encode(buffered.getvalue()).decode()

        response_data = {
            "message": "Image generated successfully",
            "timestamp": datetime.now().isoformat(),
            "prompt": input.prompt,
            "image_data": img_str
        }

        return JSONResponse(content=response_data)
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

def save_response_client(response_data):
    print("Response data keys:", response_data.keys())

    # Create a directory to store responses if it doesn't exist
    os.makedirs('responses_client', exist_ok=True)

    # Generate a unique filename based on the current timestamp
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    base_filename = f"response_{timestamp}"

    # Save the image if image_data is present
    if 'image_data' in response_data:
        image_data = base64.b64decode(response_data['image_data'])
        image = Image.open(io.BytesIO(image_data))
        image_filename = f"responses_client/{base_filename}.png"
        image.save(image_filename)
    else:
        print("Warning: No image data found in the response")
        image_filename = None

    # Save the response data as JSON
    json_filename = f"responses_client/{base_filename}.json"
    response_data_to_save = response_data.copy()
    if 'image_data' in response_data_to_save:
        del response_data_to_save['image_data']  # Remove the base64 image data before saving JSON
    if image_filename:
        response_data_to_save['image_filename'] = image_filename
    with open(json_filename, 'w') as f:
        json.dump(response_data_to_save, f, indent=2)

    return image_filename, json_filename

# Create a test client
client = TestClient(app)

# Test the /generate endpoint
response = client.post("/generate", json={"prompt": "a photo of an shih-tzu astronaut riding a horse on mars"})
print(f'Response received. Status code: {response.status_code}')
print(f'Response headers: {response.headers}')
print(f'Response content type: {response.headers.get("content-type")}')
print(f'Response body keys: {response.json().keys()}')

response_data = response.json()

print("Iterating response")
for response in response_data:
    print(f'Response {response}: {response_data[response]}')
print("*************")

image_filename, json_filename = save_response_client(response_data)
if image_filename:
    print(f"Image saved as: {image_filename}")
else:
    print("No image was saved")
print(f"Response data saved as: {json_filename}")

# To run the actual server (this will block the notebook execution):
# uvicorn.run(app, host="0.0.0.0", port=8000)

# If you want to access the API from outside Colab, use ngrok:
# !pip install pyngrok
# from pyngrok import ngrok
#
# ngrok_tunnel = ngrok.connect(8000)
# print('Public URL:', ngrok_tunnel.public_url)
# uvicorn.run(app, port=8000)
```

# Introduction to Stable Diffusion Image to Image Translation

```python
# Example 2: Image-to-image generation
from diffusers import StableDiffusionImg2ImgPipeline
from PIL import Image

def image_to_image():
    # Load the image-to-image pipeline
    pipe = StableDiffusionImg2ImgPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
    pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")

    # Load an initial image
    init_image = Image.open("input_image.png").convert("RGB")
    init_image = init_image.resize((512, 512))

    # Generate a new image based on the input image and a prompt
    prompt = "Make the season winter"
    image = pipe(prompt=prompt, image=init_image, strength=0.75, guidance_scale=7.5).images[0]

    # Save the generated image
    image.save("winter_landscape.png")

image_to_image()
```

# Introduction to Stable Diffusion Image Inpainting

```python
# Example 3: Inpainting
from diffusers import StableDiffusionInpaintPipeline
from PIL import Image, ImageDraw

def create_mask_image(width, height, shape, shape_coords, save_path):
    """
    Creates a mask image of specified dimensions with a specified shape.

    Parameters:
    width (int): Width of the mask image.
    height (int): Height of the mask image.
    shape (str): Shape to draw on the mask. Options are 'rectangle', 'ellipse', 'polygon'.
    shape_coords (tuple or list): Coordinates of the shape to draw.
    save_path (str): Path to save the created mask image.

    Returns:
    None
    """
    # Create a new image with a black background
    mask = Image.new("L", (width, height), 0)

    # Create a draw object
    draw = ImageDraw.Draw(mask)

    # Draw the specified shape
    if shape == 'rectangle':
        draw.rectangle(shape_coords, fill=255)
    elif shape == 'ellipse':
        draw.ellipse(shape_coords, fill=255)
    elif shape == 'polygon':
        draw.polygon(shape_coords, fill=255)
    else:
        raise ValueError("Invalid shape. Use 'rectangle', 'ellipse', or 'polygon'.")

    # Save the mask image
    mask.save(save_path)


def inpainting():

    # Load the inpainting pipeline
    pipe = StableDiffusionInpaintPipeline.from_pretrained("runwayml/stable-diffusion-inpainting", torch_dtype=torch.float16)
    pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")

    # Load the original image and the mask
    image = Image.open("input_image.png").convert("RGB")

    # Resize images to 512x512
    image = image.resize((512, 512))
    #mask_image = mask_image.resize((512, 512))
    create_mask_image(512, 512, 'rectangle', [100, 100, 400, 400], 'mask_image.png')
    ## create masked image
    mask_image = Image.open("mask_image.png").convert("RGB")


    # Generate the inpainted image
    prompt = "A cute cat sitting on a couch"
    image = pipe(prompt=prompt, image=image, mask_image=mask_image).images[0]

    # Save the inpainted image
    image.save("inpainted_image.png")


inpainting()
```
